```python
from scipy.signal import firwin, lfilter

#sampling properties
tlen = 3.0 #seconds
samprate = 44100 #44KHz
dt = 1.0 / samprate
t = np.arange(0.0, tlen, dt)
nyq = samprate / 2.0

print 'Nyquist Frequency: %0.0f Hz' % nyq

#construct signal as sum of sine waves
nharmonics = 5
basefreq = 440.0
freqs = [(k+1)*basefreq for k in range(nharmonics)]
s = np.zeros(len(t))
for f in freqs:
    s += np.sin(2*pi*t*f)

plt.figure()
plt.plot(t[:200], s[:200], 'k-')
plt.title('Base Signal')
plt.xlabel('Time (s)')    
plt.axis('tight')
    
#examine signal frequency properties
sft = np.fft.fft(s)
sft_freq = np.fft.fftfreq(len(s), dt)
sft_pwr = np.real(np.abs(sft))
sft_mean_pwr = sft_pwr.max()
sft_pair = np.array(zip(sft_freq, sft_pwr))
freq_index = (sft_pair[:, 0] >= 0.0) & (sft_pair[:, 0] <= np.max(freqs)+100)

plt.figure()
plt.plot(sft_pair[freq_index, 0] , sft_pair[freq_index, 1], 'k-')
plt.plot(freqs, np.ones(nharmonics)*sft_mean_pwr, 'ro')
plt.xlabel('Frequency (Hz)')
plt.axis('tight')
plt.title('Signal Power Spectrum')

#design a low pass filter
ntaps = 501
cutoff = 1500.0
filt = firwin(ntaps, cutoff, nyq=nyq)
plt.figure()
plt.plot(filt, 'r-')
plt.title('FIR Filter')

#look at power spectrum of filter
filtft = np.fft.fft(filt)
filtft_freq = np.fft.fftfreq(len(filt))*nyq
filtft_pair = np.array(zip(filtft_freq, np.abs(filtft)))
freq_index = (filtft_freq >= 0.0) & (filtft_freq <= cutoff)
plt.figure()
plt.plot(filtft_pair[freq_index, 0], filtft_pair[freq_index, 1], 'k-')
plt.title('Filter Power Spectrum')
plt.xlabel('Frequency (Hz)')

#filter the signal
sfilt = lfilter(filt, 1.0, s)
plt.figure()
plt.plot(t[:1000], s[:1000], 'k-')
plt.plot(t[:1000], sfilt[:1000], 'g-')
plt.title('Filtered Signal')
plt.legend(['Original', 'Filtered'])
plt.xlabel('Time (s)')    
plt.axis('tight')
    
#examine signal frequency properties
sft = np.fft.fft(sfilt)
sft_freq = np.fft.fftfreq(len(s), dt)
sft_pwr = np.real(np.abs(sft))
sft_mean_pwr = sft_pwr.max()
sft_pair = np.array(zip(sft_freq, sft_pwr))
freq_index = (sft_pair[:, 0] >= 0.0) & (sft_pair[:, 0] <= np.max(freqs)+100)

plt.figure()
plt.plot(sft_pair[freq_index, 0] , sft_pair[freq_index, 1], 'k-')
plt.plot(freqs, np.ones(nharmonics)*sft_mean_pwr, 'ro')
plt.xlabel('Frequency (Hz)')
plt.axis('tight')
plt.title('Filtered Signal Power Spectrum')
```
